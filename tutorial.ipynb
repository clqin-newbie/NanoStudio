# Graphene


## single-layer graphene

```python
from model import Graphene
from ase.visualize import view
model = Graphene(4, 2, -2, 6, xy_period=[1, 1], xy_pbc=[True, True])
struct = model.get_structures(layer_num=1, translations=[(0, 0)])
model.show_properties(layer_num=1)
struct.write("./examples/1.vasp", format='vasp')  
view(struct)
```

## A-B stacked SiC nanoribbons

```python
from model import Graphene
from ase.visualize import view
model = Graphene(4, 2, -2, 6, xy_period=[1, 1], xy_pbc=[True, False], atom_type=['Si', 'C'], bond_length=1.88)
struct = model.get_structures(layer_num=2, translations=[(0, 0), (1/3, 1/3)], layer_spacing=3.4)
model.show_properties(layer_num=2)
struct.write("./examples/2.vasp", format='vasp')  
view(struct)
```

## ABC-stacked h-BN nanoclusters

```python
from model import Graphene
from ase.visualize import view
model = Graphene(4, 2, -2, 6, xy_period=[1, 1], xy_pbc=[False, False], atom_type=['B', 'N'], bond_length=1.45)
struct = model.get_structures(layer_num=3, translations=[(0, 0), (1/3, 1/3), (-1/3, 2/3)])
model.show_properties(layer_num=3)
struct.write("./examples/3.vasp", format='vasp')  
view(struct)
```

# Twisted graphene

## AB Stacking twisted graphene

```python
from model import TwistGraphene
from ase.visualize import view 
model = TwistGraphene(10, 5, xy_period=[1, 1], xy_pbc=[True, True])
struct = model.get_structures(layer_num=2, translations=[(0, 0), (0, 0)], layer_sequence=[0, 1])
model.show_properties(layer_num=2)
struct.write("./examples/4.vasp", format='vasp')
view(struct)
```

## ABA Stacked twisted SiC

```python
from model import TwistGraphene
from ase.visualize import view 
model = TwistGraphene(10, 5, xy_period=[1, 1], xy_pbc=[True, True], atom_type=['Si', 'C'], bond_length=1.88)
struct = model.get_structures(layer_num=3, translations=[(0, 0), (0, 0), (0, 0)], layer_sequence=[0, 1, 0])
model.show_properties(layer_num=3)
struct.write("./examples/5.vasp", format='vasp')
view(struct)
```

# Nanotube

## single-walled carbon nanotube

```python
from model import Nanotube
from ase.visualize import view 
model = Nanotube([15], [0], z_period=[4], atom_type=['C'], z_pbc=True)
struct = model.get_structures(translation=[(0, 0)])
model.show_properties()
struct.write("./examples/6.vasp", format='vasp')
view(struct)
```

## double-walled SiC nanotubes

```python
from model import Nanotube
from ase.visualize import view 
model = Nanotube([8, 16], [2, 4], z_period=[2, 2], atom_type=['Si', 'C'], z_pbc=True, bond_length=1.88)
struct = model.get_structures(translation=[(0, 0), (0, 0)])
struct.write("./examples/7.vasp", format='vasp')
model.show_properties()
view(struct)
```

## triple-walled h-BN nanotube

```python
from model import Nanotube
from ase.visualize import view 
model = Nanotube([8, 16, 24], [2, 4, 6], z_period=[2, 2, 2], atom_type=['B', 'N'], z_pbc=True, bond_length=1.45)
struct = model.get_structures(translation=[(0, 0), (0, 0), (0, 0)])
struct.write("./examples/8.vasp", format='vasp')
model.show_properties()
view(struct)
```
